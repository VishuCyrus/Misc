```python
import pandas as pd, re, zipcodes
```

```python
pat1 = re.compile('.+[a-z]+.+')
pat2 = re.compile('^\d{1,2}$')
pat3 = re.compile('^[-+]?\d{3,4}$')
pat4 = re.compile('^[-+]?\d{9}$')
pat5 = re.compile('^[-+]?\d{5}[-+]\d{4}$')
pat6 = re.compile('^[-+]?\d{5}$')
```

```python
def zipValidation(x):
    x = str(x)
    if bool(re.match(pat1, x)): return "ZIPCODE not a number"
    if bool(re.match(pat2, x)): return "ZIPCODE must be greater than 2 numbers"
    if bool(re.match(pat3, x)): return x.replace('-','').replace('+','').zfill(5)
    if bool(re.match(pat4, x)) and x.startswith('0000'): return x.replace('-','').replace('+','')[4:]
    if bool(re.match(pat5, x)): return x.replace('-','').replace('+','')[:5]
    if bool(re.match(pat6, x)): return x.replace('-','').replace('+','')
    return "Check your ZIPCODE"
```

```python
d = pd.read_excel('dummy_zip.xlsx')
```

```python
d['res'] = d['Zip Codes'].apply(str).apply(zipValidation)
```

```python
d['zip_is_real']  = d['res'].apply(lambda x: zipcodes.is_real(x) if x.isnumeric() else False)
```
